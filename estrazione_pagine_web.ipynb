```
html_page=urlopen("https://profession.ai")
type(html_page)
html_code=html_page.read()

print(html_code)
```

```
soup= BeautifulSoup(html_code)
print(soup.prettify())
```

```
paragraphs= soup.find_all('p')
text=""

for p in paragraphs:
  text+=p.get_text()+" "

#rimuovo spazi bianchi
text=" ".join(text.split())
text=text.replace(".", "\n")
print(text)
```
